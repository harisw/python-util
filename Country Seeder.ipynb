```python
import requests as req
from bs4 import BeautifulSoup
import pandas as pd
import csv
import psycopg2
```

```python
with open("country_code_list.html") as fp:
    soup = BeautifulSoup(fp, 'html.parser')

# for city in soup.find_all('span', {'class' : 'city-sh'}):
#     print(city)
```

```python
rows = soup.find_all('tr')
countries = []
for row in rows:
    cols = row.find_all('td')
    if len(cols) < 5:
        continue
    if cols[1].find('a') != None:
        country_name = cols[1].find('a').text
    elif cols[1].find('em') != None:
        country_name = cols[1].find('em').text
    else:
        country_name = cols[1].text
    countries.append([country_name.encode('utf-8'), cols[2].text.strip(), cols[3].text.strip(), cols[4].text.strip()])
```

```python
df = pd.DataFrame(countries, columns=['name', 'a2_code', 'a3_code', 'un_code'])
```

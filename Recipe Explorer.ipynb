```python
import pandas as pd
```

```python
df = pd.read_csv('recipes_w_search_terms.csv')
```

```python
part = ['001', '002', '003', '004', '005']
for idx, val in enumerate(part):
    if val == '005':
        sliceidx = idx* 100000
        df[sliceidx:].to_csv(val+'recipes_w_search_terms.csv')
    else:
        sliceidx = (idx + 1)* 100000
        df[:sliceidx].to_csv(val+'recipes_w_search_terms.csv')
    print('done + ', val+'recipes_w_search_terms.csv')
```

```python
df = pd.read_csv('001recipes_w_search_terms.csv')
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0', 'serving_size'])
df.head()
```

```python
df['search_terms'] = df.search_terms.apply(lambda val: val.replace('{','[').replace('}',']'))
df.info()
```

```python
df.head()
```

```python
def mergeTags(first_tags, sec_tags):
    return first_tags.replace(']', '') + sec_tags.replace('[', '')

df['merged_tags'] = df.apply(lambda x: mergeTags(x.tags, x.search_terms), axis=1)
df.head()
```

```python
df = df.drop(columns=['tags', 'search_terms'])
df.head()
```

```python
df = df.sort_values('id')
df.head()
```

```python
df.isna().any()
```

```python
df.to_csv('001_preprocessedrecipes.csv', index=False)
```

```python
df = pd.read_csv('001_preprocessedrecipes.csv')
df.head()
```

```python
sample_datas = df[:1000]
sample_datas.info()
```

```python
sample_datas['tags'] = sample_datas.merged_tags.apply(lambda val: val.replace('[','').replace(']',''))
sample_datas['ingredient_tags'] = sample_datas.ingredients.apply(lambda val: val.replace('[','').replace(']',''))
sample_datas['ingredients'] = sample_datas.ingredients_raw_str.apply(lambda val: val.replace('[','').replace(']',''))
sample_datas['steps'] = sample_datas.steps.apply(lambda val: val.replace('[','').replace(']',''))
sample_datas.head()
```

```python
sample_datas = sample_datas.drop(columns=['ingredients_raw_str', 'merged_tags'])
sample_datas.head()
```

```python
sample_datas.to_csv('1k_preprocessedrecipes.csv', index=False)
```
